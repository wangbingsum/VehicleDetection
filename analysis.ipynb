```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_name = r'./output/20210828/358/DetectionResults.csv'
df = pd.read_csv(file_name)
```

```python
high = ['32', '33', '42', 'B5', 'B7'] # 32/33/42/45/B5/B7
high_package = []
low_package = []
for index, row in df.iterrows():
    vehicle_info = row.to_numpy()
    package = vehicle_info[1][7:7+2]
    if package in high:
        high_package.append(vehicle_info[2])
    else:
        low_package.append(vehicle_info[2])
print(f'high: {len(high_package)}, low: {len(low_package)}')
high_package_OK = high_package.count('OK')
high_package_NG = high_package.count('NG')
low_package_OK = low_package.count('OK')
low_package_NG = low_package.count('NG')
print(f'high ok: {high_package_OK}, high ng: {high_package_NG}, low ok: {low_package_OK}, low ng: {low_package_NG}')

plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
labels = ['high_ok','high_ng','low_ok','low_ng']
sizes = [high_package_OK, high_package_NG, low_package_OK, low_package_NG]
explode = (0, 0, 0, 0)
plt.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=False,startangle=150)
plt.title("检测结果分布统计")
plt.show(); 
```

```python
labels = ['low_ok','low_ng']
sizes = [low_package_OK, low_package_NG]
explode = (0, 0)
plt.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=False,startangle=150)
plt.title("低配检测结果分布统计")
plt.show(); 
```

```python
labels = ['high','low']
sizes = [len(high_package), len(low_package)]
explode = (0, 0)
plt.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=False,startangle=150)
plt.title("高低配分布统计")
plt.show(); 
```
